```python
request_headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.114 Safari/537.36",
    "Accept-Language": "ko-KR,ko;q=0.9,en-US;q=0.8,en;q=0.7",
    "Accept-Charset": "application/x-www-form-urlencoded; charset=UTF-8",
    "Origin": "https://developer.riotgames.com",
    "X-Riot-Token": api_key  #Riot API 키   
}
```

```python
champion_df = pd.read_csv('../00. data/champions.csv', index_col = 0)
champion_df.head()
```

```python
item_df = pd.read_csv('../00. data/items.csv', index_col = 0)
item_df.head()
```

```python
# 유저의 암호화된 아이디를 불러오는 함수 유저의 암호화된 아이디를 불러오는 함수
def searchUserId(nickname):
    user_data = requests.get("https://kr.api.riotgames.com/lol/summoner/v4/summoners/by-name/" + parse.quote(nickname), headers=request_headers).json()
    userId = user_data['id']
    userAccountId = user_data['accountId']
    userPuuid = user_data['puuid'] 
    return userId, userAccountId, userPuuid
# 한유저의 매치ID들을 불러오는 함수
def searchMatchId(puuId):
    match_list = requests.get(f"https://asia.api.riotgames.com/lol/match/v5/matches/by-puuid/{puuId}/ids?type=ranked&start=0&count=100",headers=request_headers).json()
    return match_list
# 영문 챔피언 이름으로 챔피언 이름 불러오는 함수
def searchChampion(championKey):
    champ = champion_df['name'][champion_df['name'][champion_df['eng_name'] == championKey].index[0]]
    return champ  
# 아이템키로 아이템 이름 불러오는 함수
def searchItem(itemKey):
    item = item_df['name'][item_df['name'][item_df['key'] == itemKey].index[0]]
    return item  
```

```python
def match_record(matchId):
    user_list = []
    users_champion = []
    users_kda = []
    users_items = []
    users_lane = []
    users_win =[]
    match_record = requests.get('https://asia.api.riotgames.com/lol/match/v5/matches/' + str(matchId), headers=request_headers).json()
    for user_data in match_record['info']['participants']:
        user_list.append(user_data['summonerName'])
        users_champion.append(searchChampion(user_data['championName']))
        users_lane.append(user_data['lane'])
        users_win.append(user_data['win'])
        users_kda.append(str(user_data['kills']) + '/' + str(user_data['deaths']) + '/' + str(user_data['assists']))
        user_items = []
        for i in range(6):
            itemKey = user_data['item'+str(i)]
            if itemKey == 0:
                pass
            else:
                user_items.append(searchItem(itemKey))
        users_items.append(', '.join(user_items))

    df = pd.DataFrame({
    'nickname' : user_list,
    'champion' : users_champion,
    'lane' : users_lane,
    'kda' : users_kda,
    'result_items' : users_items,
    'result' : users_win
    })
    return df
```

```python
user = '케이온'
userId, userAccountId, userPuuid = searchUserId(user)
match_list = searchMatchId(userPuuid)[:30] # 30개 불러오기
```

```python
users_champion = []
users_kda = []
users_items = []
users_lane = []
users_win =[]
for matchId in match_list:
    match_record = requests.get('https://asia.api.riotgames.com/lol/match/v5/matches/' + str(matchId), headers=request_headers).json()
    for i in range(10):
        if match_record['info']['participants'][i]['summonerName'] ==  user:
            userIndex = i
            break
    user_data = match_record['info']['participants'][userIndex]
    users_champion.append(searchChampion(user_data['championName']))
    users_lane.append(user_data['lane'])
    users_win.append(user_data['win'])
    users_kda.append(str(user_data['kills']) + '/' + str(user_data['deaths']) + '/' + str(user_data['assists']))
    user_items = []
    for i in range(6):
        itemKey = user_data['item'+str(i)]
        if itemKey == 0:
            pass
        else:
            user_items.append(searchItem(itemKey))
    users_items.append(', '.join(user_items))

df = pd.DataFrame({
'champion' : users_champion,
'lane' : users_lane,
'kda' : users_kda,
'result_items' : users_items,
'result' : users_win
})
df['trolling'] = df.apply(lambda r : True if str(r.result_items) == '' or str(r.result_items.split(', ')[0]) == str(r.result_items.split(', ')[-1]) else False, axis=1)
df
```
